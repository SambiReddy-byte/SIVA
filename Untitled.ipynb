9.LIST COMPREHNSION

```python
numbers = [1,2,3,4,5]
sqares = []

for x in numbers:
    sqares.append(x)
print(sqares)
```

```python
sqares = [x**2 for x in range(5)]
print(sqares)
```

```python
even = [x for x in range(20) if x % 2 == 0]
print(even)
```

```python
words = ["python", "machinelearning", "datascience"]
upper = [w.upper() for w in words]
print(upper)
```
